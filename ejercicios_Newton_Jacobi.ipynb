```python
#Algoritmo de Jacobi y Gráfica

#Ejercicio a
import numpy as np
from math import  * 
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

#Procedimiento :
def Jacobi(T,c,X,cotaerror,iter):
  D=np.diag(np.diag(T))
  LU=T-D
  x=X
  for i in range (iter):
    D_inv=np.linalg.inv(D)
    xtemp=x
    x=np.dot(D_inv,np.dot(-LU,x))+np.dot(D_inv, c)
    print ("Iteración",i,":x = ",x ) #Impresión de los resultados
    if np.linalg.norm(x-xtemp)<cotaerror:
      return x
  return x


#Definimos la matriz
T=np.array([
[3,-1,1],
[3,6,2],
[3,3,7]        
])
#Definimos las soluciones
c=([1,0,4])
X=np.zeros(3)  #Valores Iniciales de la Matriz
cotaerror=1e-3 #Cota de error o tolerancia
iter=500       #Número de iteraciones veces que se repetira
x=Jacobi(T,c,X,cotaerror,iter)

#Grafica de los métodos
p,f=np.linspace(0,10,10), np.linspace(0,10,10)
P,F=np.meshgrid(p,f)

Z1=(1+3*P-1*F)/1
Z2=(0+3*P+6*F)/2
Z3=(4+3*P+3*F)/7

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(P,F,Z1, alpha=0.5, cmap=cm.Accent,rstride=100,cstride=100)
ax.plot_surface(P,F,Z2, alpha=0.5, cmap=cm.Paired,rstride=100,cstride=100)
ax.plot_surface(P,F,Z3, alpha=0.5, cmap=cm.Pastel1,rstride=100,cstride=100)
ax.set_xlabel('x'); ax.set_ylabel('Y');ax.set_zlabel('Z')
plt.show()

 
```

```python
#Ejercicio a Newton

import numpy as np
from math import * 

def jacobiano(x):#derivada y obtener inversa
    J = np.array([[8*x[0]-20, 1/2*x[1]], [1/2*x[1]**2, -5]])
    JInversa = np.linalg.inv(J)
    return JInversa

def Fx(x):
    xk=np.array([4*x[0]**2-20*x[0]+1/4*x[1]**2+8, 1/2*x[0]*x[1]**2+2*x[0]-5*x[1]+8]) #matriz inicial
    return xk
x = np.array([0.1, 0.1]) #vector de valores iniciales
error = 1
c=0
while error>0.01:
    c+=1
    r= jacobiano(x)
    feval = Fx(x)
    resultado =x- np.dot(r,feval)
    
    numerador = np.amax( abs((resultado - x)))
    denominador = np.amax( abs(resultado))
    error = numerador/denominador
    x=resultado
    
    print("Iteración:",c," Resultado ",resultado)
    
```
